# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Create sample data

```python
# Sample data creation
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [200, 300, 400, 500, 600, 700],
    'Profit': [20, 30, 40, 50, 60, 70]
}

# Convert to DataFrame
df = pd.DataFrame(data)
df
```

# Pie Chart For Sales Distribution

```python
# Create a pie chart for Sales distribution
plt.figure(figsize=(8, 8))
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Sales Distribution by Month', fontsize=16)
plt.show()
```

# Histogram for Sales Data Distribution

```python
# Create a histogram for Sales data distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Sales'], bins=5, color='coral', edgecolor='black')
plt.title('Sales Data Distribution', fontsize=16)
plt.xlabel('Sales Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

# Scatter Plot for Sales vs Profit

```python
# Scatter plot for Sales vs Profit
plt.figure(figsize=(8, 6))
plt.scatter(df['Sales'], df['Profit'], color='blue', marker='o')
plt.title('Scatter Plot of Sales vs Profit', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(True)
plt.show()
```

# Box Plot for Sales

```python
# Box plot for Sales
plt.figure(figsize=(8, 6))
plt.boxplot(df['Sales'], vert=True, patch_artist=True)
plt.title('Box Plot of Sales', fontsize=16)
plt.ylabel('Sales', fontsize=12)
plt.grid(True)
plt.show()
```
